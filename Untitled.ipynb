```python
import math
#使用数学函数前必须导入模块math
```

```python
#正弦余弦曲线
#生成横坐标，从0-2pi,间隔为0.1
x= np.arange(0,2*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
#画出折线图，在同一张图上可以画多条直线
plt.plot(x,y_sin,c="r",alpha=0.2,label="sin")
plt.plot(x,y_cos,c="b",linewidth=1,label="cos")
#color线的颜色
#marker "点"的形状
#aplpha透明度
#linewidth int 线框
#linestyle str "线"的样式 "-","--",":","-.",".","none"为透明线，看不见
# 标题
plt.title("正弦曲线和余弦曲线")
#设置x和y轴的范围
plt.xlim(0,7)
plt.ylim(-2,2)
#x,y轴的标签
plt.xlabel("x")
plt.ylabel("y")
#显示图例,loc可以改变图例的显示位置
plt.legend(loc=1)
plt.show()
```

协方差和相关系数

两组数据a,b
离差分别是dev_a,dev_b
cov_ab=mean(dev_a*dev_b)

```python
import matplotlib.pyplot as plt
```

```python
a=np.array([1,2,3,5,7,7,9,10])
b=a+2
c=np.array([2,1,1,2,2,1,1,0.4])
x=np.arange(1,8+1)
```

```python
plt.plot(x,a,label="a")
plt.plot(x,b,label="b")
plt.plot(x,c,label="c")
plt.legend()
plt.show()
```

```python
#计算离差
dev_a=a-a.mean()
dev_b=b-b.mean()
dev_c=c-c.mean()
#计算标准差
std_a=a.std()
std_b=b.std()
std_c=c.std()
#计算协方差
cov_ab=np.mean(dev_a*dev_b)
cov_ac=np.mean(dev_a*dev_c)
#计算相关系数
cor_ab=cov_ab/(std_a*std_b)
cor_ac=cov_ac/(std_a*std_c)
print(cor_ab)
print(cor_ac)
```

为了消除数量集都协方差大小的影响，我们引入相关系数

相关系数的取值范围[-1,1]=cov_ab/(std_a*std_b)

#这里的std_a,std_b是标准差，非样本标准差

#编程：

y=w0 + w1x1 + w2x2 +... + wnxn

#损失函数

loss=(y-y')^2.sum()/2
      
预测未知的值（最佳逼近）


梯度下降法

向梯度相反的方向移动，每次移动一小步，组件逼近最值

优点：梯度下降无视过程，只关注结果

缺点：w型图像无法准确适用，无法解决梯度消失的问题，梯度爆炸

梯度下降法的下降选择：

1.找一个随着迭代会发生变化的参数，比如k（可能会发生梯度爆炸）

2.选择一个固定值，一般比较小0.0001    (可以防止梯度爆炸，但是效率极低，且精度与固定值无关)

3.找一个随着迭代会发生变化的参数，比如k，用学习率要约束变化

学习率/权重

约束学习步幅的参数，调参大小的参数

```python
import matplotlib.pyplot as plt
```

```python
train_x=np.array([0.5,0.6,0.8,1.1,1.4])
train_y=np.array([5.0,5.5,6.0,6.8,7.0])
n_epochs=5000#设置迭代次数
lrate=0.01#设置学习率
epoch=[]#设置列表存储的第几次迭代
losses=[]#用列表存储每次迭代的损失函数


w0,w1=[100],[100]#设置w0,w1的初始值，这个值是随意设置的，只要保证迭代次数足够多，不影响最终模型结果
for i in range (1,n_epochs+1):
    epoch.append(i)
    
    #计算预测值,是一个数组
    predict_y=train_x*w1[-1]+w0[-1]
    
    loss=((train_y-predict_y)**2).sum()/2  #计算损失函数
    
    losses.append(loss)
    
    print("这是第{}次调参，w0是{},w1是{}，本次损失函数是{}".format(i,w0[-1],w1[-1],loss))
    
    d0=(predict_y - train_y).sum()#计算w0的偏导数
    d1=(train_x*(predict_y - train_y)).sum()#计算wl的偏导数
    
    w0.append(w0[-1]-(d0*lrate))#w0的调优
    w1.append(w1[-1]-(d1*lrate))#w1的调优
```

#为什么有的数据调整很多次，损失函数还是很大

原因：

1.数据异常值很多

2.检查初始值，迭代次数，学习率

#画出迭代次数和损失函数的变化规律

```python
plt.figure("linear_model")#z指定画布
plt.plot(epoch,losses,color="r",linestyle="-.")
plt.xlim(0,20)
plt.show()
```

过拟合原因：

欠拟合：
训练集得分不高，测试集也不高
原因：模型不合适，迭代次数过少，训练集太少

分类决策树：


决策树的思想：根据特征进行层层分类，提高数据纯度，进行分类预测

H=-p(x)log2(x).sum()

P取值范围是[0,1],log2(x)取值范围是(-∞，0]，H的取值范围的[0,∞)


```python
import math
import numpy as np
import matplotlib.pyplot as plt 
```

```python
class_num=10
def get_h(n):
    p=1/n
    return -p*math.log2(p)*n
entropies=[]
for i in range (1,class_num+1):
    entropies.append(get_h(i))
plt.plot(np.arange(10),entropies,c="r",label="entropies")
plt.show()
```
